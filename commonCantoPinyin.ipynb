```python
import pandas as pd
```

```python
train=pd.read_csv('list.tsv', sep='\t', header=0)
```

```python
train["INIT"] = train["INIT"].fillna("")
train["FINL"] = train["FINL"].fillna("")
train['combine'] = train["INIT"] + train['FINL']
```

```python
newValList = train['combine'].values.tolist()
uniqueID = list(dict.fromkeys(newValList))
```

```python
uniqueID = [i.lower() for i in uniqueID if i]
uniqueID
```

```python
uniqueID.sort()
df = pd.DataFrame(uniqueID, columns=["cantoPhonics"])
df.to_csv('uniqueList.csv', index=False)
```
