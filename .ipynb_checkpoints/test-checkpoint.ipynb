```python
import fitz
from pyhanlp import HanLP
```

```python
import re
```

```python
pdf_path = './1.pdf'
doc = fitz.open(pdf_path)
```

```python
for page in doc:
    text = page.get_text()
    print(text)
```

```python
segment = HanLP.newSegment().enableNameRecognize(True);
```

```python
for page in doc:
    text = page.get_text()
    result = segment.seg(text)
    for term in result:
        if re.search(r'/nr', str(term)):
            print(str(term))
#     print(result)
```
